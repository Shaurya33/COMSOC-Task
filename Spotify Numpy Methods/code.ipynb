EXTRA METHODS

```python
#METHOD-11      np.searchsorted()   Find indices where elements should be inserted to maintain order
np.searchsorted(sorted_key, 4.5)
```

```python
#METHOD-12      np.extract()    Return the elements of an array that satisfy some condition
condition= df['key']==1
np.extract(condition, df['key'])
```

```python
#METHOD-13      np.partition()  Returns a partitioned copy of an array.
np.partition(key_arr,4)
```

```python
#METHOD-14      np.argsort()    Returns the indices that would sort an array.
np.argsort(key_arr)
```

```python
#METHOD-15      np.where()      Returns elements chosen from x or y depending on condition.
np.where(sorted_key < 5, sorted_key, 10*sorted_key)
```
